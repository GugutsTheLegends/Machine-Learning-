```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
vendas = pd.read_excel('/content/vendas.xlsx')
```

```
x = pd.get_dummies(vendas[['test']])
x.head()
```

```
y = vendas[["cliente_comprou"]]

y
```

```
from sklearn import tree
```

```
arvore = tree.DecisionTreeClassifier()
```

```
modelo = arvore.fit(x, y)
```

```
modelo
```

```
from sklearn import tree
fig = plt.figure(figsize=(70,30))

fig = tree.plot_tree(modelo,
                   feature_names=x.columns,
                   class_names=['yes', "no"],
                   filled=True)
```
